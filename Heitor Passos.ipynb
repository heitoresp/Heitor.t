```python
import pandas as pd

# Lê o arquivo CSV com delimitador ponto e vírgula
df = pd.read_csv('frota_onibus.csv', delimiter=';')

# Calcula a média dos anos de fabricação dos ônibus
media_ano_fabricacao = df['ano_fabricacao'].mean()

print(f'A média dos anos de fabricação dos ônibus é: {media_ano_fabricacao:.2f}')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Lê o arquivo CSV com delimitador ponto e vírgula
df = pd.read_csv('frota_onibus.csv', delimiter=';')

# Conta a quantidade de ônibus por ano
contagem_onibus = df['ano_fabricacao'].value_counts().sort_index()

# Gera o gráfico de barras
contagem_onibus.plot(kind='bar', figsize=(10, 6))
plt.title('Quantidade de Ônibus por Ano de Fabricação')
plt.xlabel('Ano de Fabricação')
plt.ylabel('Quantidade de Ônibus')
plt.show()
```
